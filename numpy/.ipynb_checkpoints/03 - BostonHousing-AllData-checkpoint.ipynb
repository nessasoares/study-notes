```python
test_ratio = 0.3 # tamanho do set de treinamento
shuffled_indices = np.random.permutation(len(boston_df))
test_set_size = int(len(boston_df)*test_ratio)

# √çndices do set de treinamento e teste
test_indices = shuffled_indices[:test_set_size]
train_indices = shuffled_indices[test_set_size:]

#Montagem dos sets de treinamento e teste
train_df = boston_df.iloc[train_indices]
test_df = boston_df.iloc[test_indices]
```

```python
# Gerando dados de treinamento e teste

# tenho x e quero predizer y
#Transformando em array do numpy
X_train = np.array(train_df["rm"])
Y_train = np.array(train_df["medv"])

X_test = np.array(test_df["rm"])
Y_test = np.array(test_df["medv"])
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(X_train, Y_train, c='red', alpha=0.5)
plt.xlabel("quantidade de quartos - rm")
plt.ylabel("media de pre√ßo = medv")
plt.show()
```

O problema agora √© que dado o gr√°fico, eu quero uma reta que melhor descreve a rela√ß√£o entre o <b>atributo preditor</b> rm(quantidade de quartos) e medv(media de pre√ßo)

<b>Hip√≥tese: Reta

medv = theta0 + theta1 * rm</b>

Como fazer isso? Como estimar theta0 e theta1? 

theta0 + theta1 * rm = medv


```python
theta = [0,0]
```

```python
def predict_value(theta, b):
    predicted_value = theta[0] + theta[1] * b
    return predicted_value
```

```python
def cost_function(theta, x, y):
    m = x.shape[0]
    distancia = 0
    for i in range(0, m):
        rm  = x[i]
        predicted_value = theta[0] + theta[1] * rm
        valor_real = y[i]
        
        distancia += abs(predicted_value - valor_real)
        
    cost_value = distancia / (2*m)
    return cost_value
        
    
```

```python
cost_function(theta, X_train, Y_train)
```

```python
iters = 1500

m = X_train.shape[0]
alpha = 0.3
theta = [0,0]

global_minimum = {
    'theta_0':0,
    'theta_1':0,
    'cost': 1000000000,
}

for i in range(0, iters):
    cost_0 = 0
    cost_1 = 0
    for j in range(0,m):
        predicted_value = predict_value(theta, X_train[j])
        cost_0 += (predicted_value - Y_train[j])
        cost_1 += ((predicted_value - Y_train[j]) * X_train[j])

    temp_0 = theta[0] - alpha * cost_0 *1/m
    temp_1 = theta[1] - alpha * cost_1 *1/m

    theta[0] = temp_0
    theta[1] = temp_1
    
    # Pegando o menor valor poss√≠vel 
    cost_ = cost_function(theta, X_train, Y_train)
    if cost_ < global_minimum['cost']:
        global_minimum['cost'] = cost_
        global_minimum['theta_0'] = theta[0]
        global_minimum['theta_1'] = theta[1]
        print('UPDATING BETTER COST... ')
        
        
    print("COST: {} ------ THETA 0: {}  THETA 1: {} ".format(str(cost_),str(theta[0]),str(theta[1])))

print(global_minimum)
```

```python
theta[0], theta[1]  = global_minimum['theta_0'], global_minimum['theta_1']
predicted_values = theta[0]+theta[1]* X_train
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(X_train, Y_train, c='red', alpha=0.5)
plt.plot(X_train, predicted_values,c="blue")
plt.xlabel("quantidade de quartos")
plt.ylabel("media de pre√ßo")
plt.show()
```
